Part (A) - Plot location of holes 

```python
# useful libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# import data from csv using pandas

df = pd.read_csv("code\1-Fragment_C_Hole_Measurements.csv")
```

```python
#display data
df
```

```python
# show positions of holes for each available section of the mechanism

# X-Y measurements
X = df['Mean(X)']
Y = df['Mean(Y)']

# split dataset into 8 fractured sections grouped by section ID

sections = df.groupby("Section ID")

for i, section in sections:
    print(f"\nSection ID {i}:")
    print(section)   

# can extract each section with .get_group(ID)
section_0 = sections.get_group(0)
section_1 = sections.get_group(1)
section_2 = sections.get_group(2)
section_3 = sections.get_group(3)
section_4 = sections.get_group(4)
section_5 = sections.get_group(5)
section_6 = sections.get_group(6)
section_7 = sections.get_group(7)
```

```python
# Colourblind distinguishable colours (works for me)
tableau20 = ['#332288', '#88CCEE', '#117733', '#DDCC77', '#CC6677', '#AA4499', '#44AA99', '#999933', '#882255', '#661100', '#6699CC', '#AA4466']

# Create the plot
fig,axes = plt.subplots(1,1, figsize=(16,6))

fig.set_dpi(300)

axes.scatter(X, Y, c='blue', edgecolor='black', alpha=0.7, s=10)  # Pretty scatterplot
# Label each point with its index (1 to 81)
for i in range(81):
    axes.text(X[i], Y[i]+1, str(i+1), fontsize=8, ha='right', va='bottom',)



# Customize labels and title
axes.set_xlabel('x-distance (mm)', fontsize=14)

axes.set_ylabel('y-distance (mm)', fontsize=14)


# Add a grid and adjust appearance
axes.grid(True, which='both', linestyle='--', linewidth=0.5)


# Show the plot
plt.show()
```
