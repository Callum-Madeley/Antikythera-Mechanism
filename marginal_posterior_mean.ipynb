```python
# use the marginal posterior mean and standard deviation from the MCMC chain 

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian 

def gauss(N, N_bar, sigma_N):

    return 1/((2*np.pi)**(1/2)*sigma_N) * np.exp(- (N - N_bar)**2 / (2*sigma_N**2))
```

```python
# search over hole numbers 340 - 370
N = np.linspace(340,370,31)

# isotropic model

N_bar_iso = 355.65 
sigma_N_iso = 3.73 

# radial-tangential model

N_bar_ali = 355.17
sigma_N_ali = 1.40
```

```python
# Gaussian fit to both models

isotropic = gauss(N,N_bar_iso,sigma_N_iso)
aligned = gauss(N,N_bar_ali,sigma_N_ali)
```

```python
# Plot integer bar
fig,axes = plt.subplots(1,1, figsize=(12,4))

fig.set_dpi(300)

axes.bar(N, isotropic, label='isotropic model')
axes.bar(N, aligned, label='radial-tangential model',alpha=0.7)

axes.set_xlabel('total number of holes',fontsize=12)
axes.set_ylabel('probability', fontsize=12)

# Add a grid and adjust appearance
axes.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.legend()
plt.show()
```
